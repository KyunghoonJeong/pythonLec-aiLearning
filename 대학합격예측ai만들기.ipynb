```python
# 이부분은 아래 내용 이후에 데이터 처리를 위한 과정

import pandas as pd
import numpy as np
# 만약 임폴트가 안되는 경우가 생기면 아나콘다 프롬프트를 관리자권한으로 들어가서 activate tensorflow로 들어간 뒤에 pip install 이름 //해줄 것 !!!!!!!

data = pd.read_csv('gpascore.csv') #csv 형식의 파일을 읽겠음 / 판다스를 쓰는이유 !

#print(data)
 
# 데이터를 가져오고 확인했으니 데이터 전처리를 하자 !!!!!!

#데이터 전처리를 왜하냐 ...? => 빵꾸난 데이터 어떻게 할거니 ????  보통 두가지 방법 => 1. 빈곳에 평균값을 넣거나  2. 아예 그 행을 삭제

#데이터 구멍난곳 확인하는 방법
#print(data.isnull().sum())  # gre라는 열에 하나가 비어있는 것 같대 !

# 코드를 통해 구멍난 곳 지울 수 있다. 
data = data.dropna()  # dropna()는 NaN/빈값있는 행을 제거해준다 ! 
#data.fillna(100) 빈칸이 있으면 그값에 100을 넣어주겠다 ! 


#print(data.isnull().sum())  #빈칸 지워진 것을 확인할 수 있다.

#print(data['gpa']) # gpa라는 열 다 볼 수도 있다. 

#유용한 기능  data['gpa'].min() 하면 그 열에서 최소값 , max() 최대값 , count() 개수 세줌.


#이제 데이터 정리를 해보자 0과 1부터 정리해보자

y데이터 = data['admit'].values  #y데이터에 값 담기 성공

#print(y데이터)

x데이터 = [] 
#반복문을 통해 넣기
for i,rows in data.iterrows():   #iterrow()는 판다스의 함수인데 데이터라는 데이터프레임을 가로 한줄씩 출력해주는 기능 (그냥 아래로 쭉 출력해주는 편리한 기능 => rows에 담김)
    #print(rows['gre']) #'gre'세로 열에 있는 데이터 출력해준다 
    #print([ rows['gre'], rows['gpa'], rows['rank'] ])  #이렇게 하면 우리가 원하는 형식이 가능 ! 이걸 이제 append해서 넣자 !
    x데이터.append([ rows['gre'], rows['gpa'], rows['rank'] ])

print(x데이터)  #우리가 원하는 방식으로 만들어졌다 ! 

```

```python
import tensorflow as tf  

#케라스를 쓰면 w값 직접 노가다 해야하지만 keras를 사용하면 쉽게 모델링 가능

#1. 딥러닝 model 디자인하기 
model = tf.keras.models.Sequential([ 
    tf.keras.layers.Dense(64, activation='tanh'),   #히든 레이어의 개수   => 레이어 개수나 노드 개수는 맘대로 해도된다. 결과가 잘 나올때까지 실험적으로 하는것
    tf.keras.layers.Dense(128, activation='tanh'),
    tf.keras.layers.Dense(1, activation='sigmoid'),   #마지막 레이어만 잘 신경쓰면 된다 ! 합격 했냐 안했냐로 나누기 때문에 1
    # activation에는 활성함수를 넣어야한다 여기엔 많은 종류가 있다. sigmoid, tanh, relu, softmax, leakyRelu 등 많이 쓰임
    # 마지막 레이어는 예측 결과를 뱉어야함. 0~1 사이의 확률을 뱉고 싶으면 sigmoid를 써보자 (시그모이드 함수는 모든 값을 0에서 1사이로 압축 !)
    # 마지막 출력결과가 0.2이면 입학확률이 0.2 인거임
    
    # 여기 부분에서도 성능을 좌지우지할 사항들이 많음. 여러 연구가 필요한 부분이다 !  
])
#2. 모델에 컴파일 해줘야 모델이 완성
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
#옵티마이저 목록 : adam, adagrad, adadelta, rmsprop, sgd 등
# 로스함수 중 바이너리 크로스엔트리 함수는 결과가 0과 1사이 분류/확률 문제에서 사용한다.
# 메트릭스는 딥러닝 할때 어떤 요소로 평가할 것인지 !

#3. 모델 학습 시키기
#model.fit(x데이터, y데이터,epochs=100 )  x데이터는 학습데이터 y데이터는 실제 출력될 값, 에포크는 몇번 학습시킬지 !!

model.fit(np.array(x데이터),np.array(y데이터),epochs=1000)

# 여기까지가 모델을 완성하였고 그뒤로는 데이터를 잘 처리하여 필요한 곳에 넣어주면 된다.

# x 데이터 값 넣을 때 []리스트 형식으로 값을 넣어주면 된다. 
# 이제 데이터를 잘 묶어서 x = [ [380,3.21,3], [660,3.67,3],[],[]~ ] 이런 형태로 넣어줘야하는데 그것은 이제 위에서 !


#x,y 값이  numpy array나 tf tensor 자료형으로 바뀌어야 학습 가능 !
 # np.array(x데이터) 넘파이 어레이로 변환시켜주는 고마운 함수다 !

# epoch이 시도마다 이루어지는 것들 나오고
# loss 는 손실값을 의미 => 적어야함 .
# acuuracy는 예측한 값이 얼마나 정확히 맞는지 ! 

#10번 할때는 정확도가 0.5넘기 어려웠는데 1000번 시키니까 80퍼가까이로 오른다.. 매우신기.....
```
