#### 문자열 사용방법

```python
# 문자열 안에 문자 '', ""추가 시키고 싶을 때
# 문자열 조합 : 이스케이프 코드 => 백슬래시 넣어주면 된다.
s = 'she\'s gone.'
print(s)
```

```python
# 개행이 되어지는 문장을 정의할 때 사용하는 문자열 기호
# '''문장''' ,  """문장(엔터그대로 적용하면 그대로 반영)"""

'안녕하세요 ~ '
'저는 손지영입니다.'

#이걸 따옴표 3개를 쓰면 개행을 알아서 인식해준다
print( '''안녕하세요 ~
저는 손지영입니다.''')
```

#### 문자열 인덱싱/ 슬라이싱

```python
s = 'My name is MH'
```

```python
print(s[3])
print(s[8])
print(s[-2])
print(s[-1])
```

```python
print(s[0:7])
print(s[0:6])
#0부터 5까지 인걸 주의 ! 
print(s[:5])  #0부터면 0 생략 가능 !
print(s[5:])  #마지막까지도 뒷부분 생략 가능 !
print(s[:])   # 전부다 출력
```

```python
day = "2020년 3월 3일의 날씨는 맑음입니다."
print("날짜 :", day[:11])
print("날씨 :", day[17:19])
```

```python
day = 25
month =10 
s = '오늘은 '+str(month) +'월 '+str(day)+'일 입니다.'
print(s)
```

```python
# %기호 포매팅

month = 10 
day = 25 

s = '오늘은 %d월 %d일 입니다.'%(month, day)    # %기호 포매팅을 활용하는 방법! 
print(s)
```

```python
# format함수 사용 포매팅 
month = 10 
day = 25 

s = '오늘은 {}월 {}일 입니다.'.format(month,day)  #중괄호를 이용해준다.
print(s)
```

```python
# f 포매팅

month = 10 
day =25 
s= f'오늘은 {month}월 {day}일 입니다.'
print(s)
```

```python
x= 100
y= 200
sum2 =x+y
s = f'{x}와 {y}의 합은 {sum2}입니다.'
print(s)
```

```python
num1 =23
num2 =3 
print(f'더하기 결과 : {num1+num2}')
print(f'빼기 결과 : {num1-num2}')
print(f'곱하기 결과 : {num1*num2}')
print(f'나누기 결과 : {num1/num2}')
```
