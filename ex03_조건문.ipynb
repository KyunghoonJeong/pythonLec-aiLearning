### 리스트 list, 튜플 Tuple

- 리스트

```python
#리스트란?
#파이썬 자료 구조 중 하나
#순서가 있는 수정가능한 객체 중 하나 
#대괄호([])로 작성되어지며, 내부의 요소는 (,)로 구분
#추가, 삭제, 수정 가능하다.
```

```python
# 리스트 정의 
a = [] # 빈 리스트 선언
b = [1,2,3] #정수 데이터 담은 리스트 
c = ['my','name','is','jy'] #문자열 데이터 담은 리스트 
d = [1,2,'my','name'] #정수, 문자열 같이 담은 리스트 
e = [1,2,['my','name']] #리스트 안에 리스트 들어갈 수 있다. (이중리스트)

```

```python
print(a)
print(b)
print(c)
print(d)
print(e)
```

- 리스트 인덱싱, 슬라이싱

```python
# 인덱싱 : 무엇인가를 가리킴
# 슬라이싱 : 무엇인가를 잘라냄(범위)
```

```python
list1 = [2,5,7,9,10]
print(list1[0])  # 2라는 숫자 
print(list1[3])  # 9라는 숫자 인덱싱 
print(list1[0]+list1[3])
```

```python
list2 = [['my'],['name'],['jy']]
print(list2[0][0],list2[1][0],list2[2][0])
```

- 리스트 슬라이싱

```python
# 리스트[시작인덱스번호:끝인덱스번호]
# 끝 인섹스 번호 값은 포함하지 않는다 .

list3 =[100,90,50,70,35]
print(list3[1:4])
print(list3[2:])
print(list3[:4])
print(list3[:])
print(list3[::2])   #0부터 끝까지 2칸씩 나타냄

# 리스트 뒤에서 읽기 
print(list3[::-1])
```

- 리스트 관련 함수

```python
# 리스트 맨 뒤로 값 추가 함수
# append(값)

lst = [] #빈 리스트 선언 
lst.append('광현')
print(lst)
```

```python
lst.append('도현')  # 어펜드 안에는 한가지 요소만 들어가야한다 ! 리스트로 묶으면 여러개 넣는것도 가능 
print(lst)
```

```python
# 사용자가 지정하는 인덱스 위치에 값을 추가 함수(기능)
# insert(인덱스 번호, 값)

lst.insert(1,'진태')
print(lst)
```

```python
lst = ['딸기','바나나','수박','체리','포도']
fruit = input("검색할 문자를 입력하세요 >> ")
if fruit in lst:
    print(f'{fruit}는 리스트에 {lst.index(fruit)}번째 인덱스에 들어있습니다')
else:
    print(f'{fruit}는 리스트에 들어있지 않습니다.')
```
